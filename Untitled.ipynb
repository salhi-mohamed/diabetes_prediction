```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df=pd.read_csv("diabetes.csv")
df
```

```python
percentage_zeros = (df.isin([0]).sum() / len(df)) * 100

# Display the result
print(percentage_zeros)
```

```python
columns_to_check = ['BMI', 'Glucose', 'BloodPressure']

# Plot boxplots for the selected columns
df[columns_to_check].plot(kind='box', subplots=True, layout=(1, 3), figsize=(15, 5), sharey=False, grid=True)
plt.tight_layout()
plt.show()
```

```python
# Filter out rows where either 'Glucose' or 'Insulin' is zero
filtered_df = df[(df['Glucose'] != 0) & (df['Insulin'] != 0)]

# Plotting the scatter plot with a line of best fit
plt.figure(figsize=(8, 6))
sns.regplot(x='Glucose', y='Insulin', data=filtered_df, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Linear Correlation between Glucose and Insulin (Without Zeros)')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
```

```python
correlation_matrix = filtered_df.corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of filtred df')
plt.show()
```

```python
# Compute the Spearman correlation matrix
spearman_corr = filtered_df.corr(method='spearman')

# Plotting the Spearman correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Spearman Correlation Matrix of filtred df')
plt.show()
```

```python
filtered_df[['Glucose','Insulin']]
```

```python
from sklearn.ensemble import RandomForestRegressor

# Define the independent variable (X) and the dependent variable (y)
X = filtered_df[['Glucose']]  # Features (Glucose values)
y = filtered_df['Insulin']   # Target (Insulin values)

# Create a Random Forest Regressor model
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model on the entire dataset
model_rf.fit(X, y)

# Predict insulin values using the trained model
y_pred_rf = model_rf.predict(X)

# Plotting the scatter plot with the regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', s=50, label='Actual Insulin values')
plt.scatter(X, y_pred_rf, color='red', s=50, label='Predicted Insulin values (RF)')
plt.title('Random Forest Regressor: Predicting Insulin from Glucose')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.legend()
plt.show()

# Print the feature importance (to understand the model's decision-making)
print(f'Feature importance: {model_rf.feature_importances_}')
```

```python
# Find the rows where Insulin is zero
zeros_df = df[df['Insulin'] == 0]
df['Old_Insulin']=df['Insulin'].copy()
# Use the glucose values from those rows to predict the insulin values
predicted_insulin_zeros = model_rf.predict(zeros_df[['Glucose']])
# Fill the zeros in the 'Insulin' column with the predicted values
df.loc[df['Insulin'] == 0, 'Insulin'] = predicted_insulin_zeros
```
